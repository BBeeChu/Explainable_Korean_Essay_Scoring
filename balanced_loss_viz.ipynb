```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
test_prompt = ['    ', '  ', ' ', '  ',
                'e    ', '     ']

for prompt in test_prompt:
    exp_values = pd.read_pickle('./result/cross_prompt/{}_exp_classwise.pkl'.format(prompt))
    org_values = pd.read_pickle('./result/cross_prompt/{}_org_classwise.pkl'.format(prompt))
    cont_values = pd.read_pickle('./result/cross_prompt/{}_cont_classwise.pkl'.format(prompt))

    fig, axs = plt.subplots(3, 1, figsize=(12, 12))

    # x  
    classes = ['class 1', 'class 2', 'class 3', 'class 4', 'class 5']

    #     
    axs[0].bar(classes, exp_values, color='blue', alpha=0.7)
    axs[0].set_title('Histogram of Exp by Class ({}) / Mean Class-wise MAE: {:.4f}'.format(prompt, np.mean(exp_values)), fontsize=20)

    axs[1].bar(classes, org_values, color='green', alpha=0.7)
    axs[1].set_title('Histogram of Org by Class ({}) / Mean Class-wise MAE: {:.4f}'.format(prompt, np.mean(org_values)), fontsize=20)

    axs[2].bar(classes, cont_values, color='red', alpha=0.7)
    axs[2].set_title('Histogram of Cont by Class ({}) / Mean Class-wise MAE: {:.4f}'.format(prompt, np.mean(cont_values)), fontsize=20)

    # x y  
    for ax in axs:
        ax.set_xlabel('Classes')
        ax.set_ylabel('Values')

    plt.tight_layout()  #    
    plt.show()
```
